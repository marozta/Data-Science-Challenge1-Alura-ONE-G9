#1. An√°lisis de facturaci√≥n



###**Ingreso anual total por tienda.**


En este primer an√°lisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Lista con los DataFrames correspondientes a cada tienda.
tiendas_df = [tienda1, tienda2, tienda3, tienda4]

# Lista con los nombres de cada tienda.
tiendas_nombre = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# C√°lculo del ingreso total de cada tienda.
total_tiendas = [tienda['Precio'].sum() for tienda in tiendas_df]

# Creaci√≥n del DataFrame con los resultados.
df_tiendas = pd.DataFrame({
    'Tiendas': tiendas_nombre,
    'Total': total_tiendas
})
```

```python
# Ingreso total por tienda
ingreso_tienda_1 = tienda1['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

# Resultados
print(f'Ingreso total tienda 1: ${ingreso_tienda_1:,.2f}')
print(f'Ingreso total tienda 2: ${ingreso_tienda_2:,.2f}')
print(f'Ingreso total tienda 3: ${ingreso_tienda_3:,.2f}')
print(f'Ingreso total tienda 4: ${ingreso_tienda_4:,.2f}')
```

# 2. Ventas por categor√≠a

En este paso se debe calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda.

###**Cantidad anual de ventas por categoria por tienda.**


```python
# Ventas por categor√≠a
ventas_categoria_tienda1 = tienda1.groupby('Categor√≠a del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda2 = tienda2.groupby('Categor√≠a del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda3 = tienda3.groupby('Categor√≠a del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda4 = tienda4.groupby('Categor√≠a del Producto')['Producto'].count().sort_values(ascending=False)

# Resultados
print("\nVentas por categor√≠a - Tienda 1:")
print(ventas_categoria_tienda1)

print("\nVentas por categor√≠a - Tienda 2:")
print(ventas_categoria_tienda2)

print("\nVentas por categor√≠a - Tienda 3:")
print(ventas_categoria_tienda3)

print("\nVentas por categor√≠a - Tienda 4:")
print(ventas_categoria_tienda4)
```

```python
# Productos por categor√≠a
cat1 = tienda1.groupby('Categor√≠a del Producto')['Producto'].count()
cat2 = tienda2.groupby('Categor√≠a del Producto')['Producto'].count()
cat3 = tienda3.groupby('Categor√≠a del Producto')['Producto'].count()
cat4 = tienda4.groupby('Categor√≠a del Producto')['Producto'].count()

# Obtener todas las categor√≠as √∫nicas
todas_categorias = sorted(set(cat1.index) | set(cat2.index) | set(cat3.index) | set(cat4.index))

```

# 3. Calificaci√≥n promedio de la tienda


En este paso, se debe calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacci√≥n del cliente con los productos vendidos.

```python

# Calcular el promedio de calificaciones por tienda
promedio_1 = tienda1['Calificaci√≥n'].mean()
promedio_2 = tienda2['Calificaci√≥n'].mean()
promedio_3 = tienda3['Calificaci√≥n'].mean()
promedio_4 = tienda4['Calificaci√≥n'].mean()

# Mostrar resultados
print(f'Promedio calificaci√≥n Tienda 1: {promedio_1:.2f} ‚≠ê')
print(f'Promedio calificaci√≥n Tienda 2: {promedio_2:.2f} ‚≠ê')
print(f'Promedio calificaci√≥n Tienda 3: {promedio_3:.2f} ‚≠ê')
print(f'Promedio calificaci√≥n Tienda 4: {promedio_4:.2f} ‚≠ê')
```

# 4. Productos m√°s y menos vendidos

En este paso, se debe identificar qu√© productos fueron los m√°s vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qu√© productos destacaron en ventas en cada tienda.

```python
def listar_productos_mas_vendidos(df, nombre_tienda):
    print(f'\nüì¶ Productos m√°s vendidos - {nombre_tienda}')

    ventas = df['Producto'].value_counts()

    # Imprimir con formato
    for producto, cantidad in ventas.items():
        print(f'‚Ä¢ {producto}: {cantidad} ventas')

# Ejecutar para cada tienda
listar_productos_mas_vendidos(tienda1, "Tienda 1")
listar_productos_mas_vendidos(tienda2, "Tienda 2")
listar_productos_mas_vendidos(tienda3, "Tienda 3")
listar_productos_mas_vendidos(tienda4, "Tienda 4")


```

```python
def listar_productos_menos_vendidos(df, nombre_tienda):
    print(f'\nüìâ Productos menos vendidos - {nombre_tienda}')

    ventas = df['Producto'].value_counts().sort_values(ascending=True)

    # Imprimir con formato
    for producto, cantidad in ventas.items():
        print(f'‚Ä¢ {producto}: {cantidad} ventas')

listar_productos_menos_vendidos(tienda1, "Tienda 1")
listar_productos_menos_vendidos(tienda2, "Tienda 2")
listar_productos_menos_vendidos(tienda3, "Tienda 3")
listar_productos_menos_vendidos(tienda4, "Tienda 4")


```

# 5. Costo env√≠o promedio por tienda

En este paso, se debe calcular el costo de env√≠o promedio para cada tienda. El objetivo es comprender cu√°nto se gasta, en promedio, en el env√≠o de cada tienda.  Sabiendo que el cliente paga el env√≠o.

```python
# C√°lculo promedio del costo de env√≠o por tienda
promedio_envio_tienda_1 = round(tienda1['Costo de env√≠o'].mean())
promedio_envio_tienda_2 = round(tienda2['Costo de env√≠o'].mean())
promedio_envio_tienda_3 = round(tienda3['Costo de env√≠o'].mean())
promedio_envio_tienda_4 = round(tienda4['Costo de env√≠o'].mean())

# Resultados
print(f'Promedio env√≠o Tienda 1: ${promedio_envio_tienda_1}')
print(f'Promedio env√≠o Tienda 2: ${promedio_envio_tienda_2}')
print(f'Promedio env√≠o Tienda 3: ${promedio_envio_tienda_3}')
print(f'Promedio env√≠o Tienda 4: ${promedio_envio_tienda_4}')
```

```python
# C√°lculo del costo total de env√≠o por tienda
total_envio_tienda_1 = round(tienda1['Costo de env√≠o'].sum())
total_envio_tienda_2 = round(tienda2['Costo de env√≠o'].sum())
total_envio_tienda_3 = round(tienda3['Costo de env√≠o'].sum())
total_envio_tienda_4 = round(tienda4['Costo de env√≠o'].sum())

# Resultados
print(f'Promedio env√≠o Tienda 1: ${total_envio_tienda_1}')
print(f'Promedio env√≠o Tienda 2: ${total_envio_tienda_2}')
print(f'Promedio env√≠o Tienda 3: ${total_envio_tienda_3}')
print(f'Promedio env√≠o Tienda 4: ${total_envio_tienda_4}')
```

###**Ingresos netos por tienda.**


```python
# Ingreso total por tienda
ingreso_tienda_1 = tienda1['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

# Costo total por tienda
total_envio_tienda_1 = round(tienda1['Costo de env√≠o'].sum())
total_envio_tienda_2 = round(tienda2['Costo de env√≠o'].sum())
total_envio_tienda_3 = round(tienda3['Costo de env√≠o'].sum())
total_envio_tienda_4 = round(tienda4['Costo de env√≠o'].sum())

# Ventas netas por tienda
ventas_netas_tienda_1 = ingreso_tienda_1 - total_envio_tienda_1
ventas_netas_tienda_2 = ingreso_tienda_2 - total_envio_tienda_2
ventas_netas_tienda_3 = ingreso_tienda_3 - total_envio_tienda_3
ventas_netas_tienda_4 = ingreso_tienda_4 - total_envio_tienda_4


# Resultados
print(f'Ingresos netos tienda 1: ${ventas_netas_tienda_1:,.2f}')
print(f'Ingresos netos tienda 2: ${ventas_netas_tienda_2:,.2f}')
print(f'Ingresos netos tienda 3: ${ventas_netas_tienda_3:,.2f}')
print(f'Ingresos netos tienda 4: ${ventas_netas_tienda_4:,.2f}')
```

##**Gr√°fico de la cantidad de ventas por categor√≠a**

```python
import matplotlib.colors as mcolors

# Lista de colores personalizada
tus_nuevos_colores = [
    '#FFB6C1',  # Rosa Claro
    '#B5B9FF',  # Lavanda
    '#A8E6CF',  # Menta Suave
    '#FFDAB9',  # Melocot√≥n
    '#ADD8E6',  # Azul Cielo Claro
    '#FFFACD',  # Amarillo Crema
    '#E6E6FA',  # Lila P√°lido
    '#FFA07A'   # Salm√≥n Claro
]

color_list = tus_nuevos_colores

# Asegurate de que todas_categorias exista antes
# Ejemplo: todas_categorias = ['Bebidas', 'L√°cteos', 'Snacks', ...]
colores_categoria = {cat: color_list[i % len(color_list)] for i, cat in enumerate(todas_categorias)}

# Funci√≥n para graficar pie chart por tienda
def graficar_torta(datos, titulo):

    # Reordenar los datos para que incluyan todas las categor√≠as
    datos_completos = [datos.get(cat, 0) for cat in todas_categorias]
    colores = [colores_categoria[cat] for cat in todas_categorias]

    # Funci√≥n personalizada: % + unidades
    def etiqueta(pct, valores):
        total = sum(valores)
        unidades = int(round(pct * total / 100.0))  # c√°lculo de unidades
        if pct < 0.1:  # evita texto en categor√≠as muy peque√±as
            return ""
        return f'{pct:.1f}%\n({unidades})'

    plt.figure(figsize=(6, 6))
    plt.pie(
        datos_completos,
        labels=todas_categorias,
        colors=colores,
        autopct=lambda pct: etiqueta(pct, datos_completos),
        startangle=140
    )

    plt.title(titulo)
    plt.axis('equal')  # Proporci√≥n 1:1
    plt.tight_layout()
    plt.show()

# Crear gr√°ficos
graficar_torta(cat1, 'Tienda 1 - Ventas por Categor√≠a')
graficar_torta(cat2, 'Tienda 2 - Ventas por Categor√≠a')
graficar_torta(cat3, 'Tienda 3 - Ventas por Categor√≠a')
graficar_torta(cat4, 'Tienda 4 - Ventas por Categor√≠a')


```

##**Gr√°fico de distribuci√≥n de valoraciones por tienda**

```python
# Crear una figura con 4 subgr√°ficos (uno por tienda)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Distribuci√≥n de calificaciones por tienda', fontsize=16)

# Datos para graficar
tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
axes = axs.flatten()

for i in range(4):

    # El total es la suma de todos los "counts" (frecuencias) en la tienda actual.
    total_calificaciones = len(tiendas[i]['Calificaci√≥n'])

    # Crear histograma y capturar barras + valores
    counts, bins, patches = axes[i].hist(
        tiendas[i]['Calificaci√≥n'],
        bins=range(1, 7),
        edgecolor='black',
        color='#CEEAD6',
        align='left'
    )

    axes[i].set_title(nombres[i])
    axes[i].set_xlabel('Calificaci√≥n')
    axes[i].set_ylabel('Frecuencia')
    axes[i].set_xticks(range(1, 6))

    max_count = max(counts)
    axes[i].set_ylim(0, max_count + 200)

    # Agregar texto arriba de cada barra
    for count, patch in zip(counts, patches):
        if count > 0:  # No etiquetar barras vac√≠as
            x = patch.get_x() + patch.get_width() / 2
            y = patch.get_height()

    # Calcular porcentaje
            porcentaje = (count / total_calificaciones) * 100

    # Formatear el texto: Cantidad \n (Porcentaje)
            texto_etiqueta = f"{int(count)}\n({porcentaje:.1f}%)" # .1f para un decimal

    # Agregar la etiqueta de texto combinada
            axes[i].text(
                x,
                y + 0.2,       # separacion del tope
                texto_etiqueta, # Usamos el texto combinado
                ha='center',
                va='bottom',
                fontsize=9    # Reducido un poco para que quepa bien
            )

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
